```
gpt2.download_gpt2(model_name='345M' )
```

```
gpt2.mount_gdrive()
```

```
file_name = "20novels.rtf"
```

```
gpt2.copy_file_from_gdrive(file_name)
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              learning_rate=1e-5,
              dataset=file_name,
              model_name='345M',
              steps= 3500 ,
              restore_from='fresh',
              print_every=1000,
              sample_every=500,
            save_every=500
              )
```

```
gpt2.finetune(sess,
              learning_rate=1e-5,
              dataset=file_name,
              model_name='345M',
              steps= 3500 ,
              restore_from='latest',
              print_every=1000,
              sample_every=500,
              run_name='run1',
            save_every=500
              )
```

```
sess = gpt2.start_tf_sess()
```

```
gpt2.copy_checkpoint_to_gdrive(run_name='run1' )
```

```
gpt2.copy_checkpoint_from_gdrive(run_name='run1')
```

```
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='run1')
```

```
gpt2.generate(sess,
              length=500,
              temperature=0.9,
              nsamples=1,
              batch_size=1,
              top_k=20,
              include_prefix=True,
              prefix="Group of friends walking down the street"
              )
```
